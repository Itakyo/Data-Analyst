## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df_cv = pd.DataFrame(data=df['ConvertedComp'])
df_cv.plot(kind='box', figsize=(8,6))
plt.title('ConvertedComp Box Plot')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
print('Primer Cuartil', Q1)
Q3 = df['ConvertedComp'].quantile(0.75)
print('Tercer Cuartil', Q3)
IQR = Q3 - Q1
print('Rango intercuartil', IQR)
Mediana = df['ConvertedComp'].median()
print('Mediana', Mediana)
Val_Minimo = df['ConvertedComp'].min()
print('Valor minimo', Val_Minimo)
Val_Maximo = df['ConvertedComp'].max()
print('Valor màximo', Val_Maximo)
```

Find out the upper and lower bounds.


```python
# your code goes here
Bigote_I = (Q1 - 1.5 * IQR)
print('Bigote inferior: ', int(Bigote_I))
Bigote_S = (Q3 + 1.5 * IQR)
print('Bigote superior: ', int(Bigote_S))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
ubicacion_outliers = (df['ConvertedComp'] < Bigote_I) | (df['ConvertedComp'] > Bigote_S)
print('Ubicacion de outliers', ubicacion_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
ubicacion_sin_outliers = (df['ConvertedComp'] >= Bigote_I) & (df['ConvertedComp'] <= Bigote_S)
sin_outliers = df[ubicacion_sin_outliers]
print(sin_outliers)
```

```python
plt.title('sin_outliers')
plt.boxplot(sin_outliers['ConvertedComp'])
plt.show()
```

```python
sin_outliers.describe()
```

```python
df_cv = pd.DataFrame(data=df['Age'])
df_cv.plot(kind='box', figsize=(8,6))
plt.title('Age Box Plot')
plt.show()
```

```python
Q1 = df['Age'].quantile(0.25)
print('Primer Cuartil', Q1)
Q3 = df['Age'].quantile(0.75)
print('Tercer Cuartil', Q3)
IQR = Q3 - Q1
print('Rango intercuartil', IQR)
Mediana = df['Age'].median()
print('Mediana', Mediana)
Val_Minimo = df['Age'].min()
print('Valor minimo', Val_Minimo)
Val_Maximo = df['Age'].max()
print('Valor màximo', Val_Maximo)
```

```python
Bigote_I = (Q1 - 1.5 * IQR)
print('Bigote inferior: ', int(Bigote_I))
Bigote_S = (Q3 + 1.5 * IQR)
print('Bigote superior: ', int(Bigote_S))
```

```python
ubicacion_outliers = (df['Age'] < Bigote_I) | (df['Age'] > Bigote_S)
print('Ubicacion de outliers', ubicacion_outliers)
```
